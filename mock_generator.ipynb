```python
import pandas as pd
import numpy as np
```

```python
def generator():
    list_of_lines = []


    for i in range(100):

        list_of_lines.append(f"Question {i+1}")

        random_ans = np.random.randint(1, 5)

        answer_texts = []
        ans1 = (f"Answer {i+1}.1")
        ans2 = (f"Answer {i+1}.2")
        ans3 = (f"Answer {i+1}.3")
        ans4 = (f"Answer {i+1}.4")

        answer_texts = [ans1, ans2, ans3, ans4]

        for index, item in enumerate(answer_texts):
            if random_ans == index+1:
                item = "[x] " + item
            else:
                item = "[ ] " + item

            list_of_lines.append(item)


    return list_of_lines
        
```

```python
for i in range(40):
    name = f"mock_{i+1}.txt"
    new_file = open(f"mock/M_data/{name}", 'w')
    generated = generator()
    for item in generated:
        new_file.write(item + "\n")
    new_file.close()
```
