## Function 1: extract_answers_sequence()

The function `extract_answers_sequence()` reads one raw quiz response file.

It goes through the file line by line.

For each question, it checks the four answer options.

If the first option has `[x]`, the answer is saved as `1`.

If the second option has `[x]`, the answer is saved as `2`.

If the third option has `[x]`, the answer is saved as `3`.

If the fourth option has `[x]`, the answer is saved as `4`.

If no option has `[x]`, the answer is saved as `0`.

At the end, the function returns a list of 100 numbers.

## Function 2: write_answers_sequence()
